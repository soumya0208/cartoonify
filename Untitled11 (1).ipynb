```python
import cv2,time
```

```python
import easygui
import numpy as np
import imageio
import sys
import matplotlib.pyplot as plt
import os
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from tkinter import messagebox
from PIL import ImageTk, Image
```

```python
imagepath=easygui.fileopenbox()
```

```python
oimage=cv2.imread(imagepath)
oimage=cv2.cvtColor(oimage,cv2.COLOR_BGR2RGB)
if oimage is None:
    print("cannot find image")
    sys.exit()
gsize=cv2.resize(oimage,(960,540))
plt.imshow(gsize,cmap='gray')
```

```python
gimage=cv2.cvtColor(oimage,cv2.COLOR_BGR2GRAY)
gsize1=cv2.resize(gimage,(960,540))
plt.imshow(gsize1,cmap='gray')
```

```python
sgray=cv2.medianBlur(gsize1,7)
gsize2=cv2.resize(oimage,(960,540))
plt.imshow(gsize2,cmap='gray')
```

```python
gedge=cv2.adaptiveThreshold(sgray,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,7,7)
gsize3=cv2.resize(gedge,(960,540))
plt.imshow(gsize3,cmap='gray')
```

```python
onimage=oimage
for i in range(5):
    data=np.float32(onimage).reshape((-1,3))
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 20, 0.001) 
    ret, label, center = cv2.kmeans(data, 9, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
    center = np.uint8(center)
    result = center[label.flatten()] 
    onimage = result.reshape(oimage.shape)
cimage=cv2.bilateralFilter(onimage,7,150,150)
gsize4=cv2.resize(cimage,(960,540))
plt.imshow(gsize4,cmap='gray')
```

```python
cartoon = cv2.bitwise_and(cimage, cimage,gedge)
gsize5=cv2.resize(cartoon,(960,540))
plt.imshow(gsize5,cmap='gray')
```

```python
images=[gsize,gsize1,gsize2,gsize3,gsize4,gsize5]
fig, axes = plt.subplots(3,2, figsize=(8,8), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1, wspace=0.1))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i], cmap='gray')
plt.show()
```

```python
newName="cartoonified_Image"
path1 = os.path.dirname(imagepath)
extension=os.path.splitext(imagepath)[1]
path = os.path.join(path1, newName+extension)
cv2.imwrite(path, cv2.cvtColor(gsize5, cv2.COLOR_RGB2BGR))
I = "Image saved by name " + newName +" at "+ path
tk.messagebox.showinfo(message=I)
```
